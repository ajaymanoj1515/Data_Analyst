```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x="Pclass",y="Survived",data=df1)
plt.title("Survived by Class")
plt.show()
```

```python
sns.countplot(x="Sex",data=df1)
plt.title("Survived by Gender")
plt.show()
```

```python
sns.histplot(df1["Age"],bins=20, kde=True)
plt.show()
```

```python
sns.heatmap(df1.corr(numeric_only=True),annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.scatterplot(x="Age",y="Fare",data=df1)
plt.show()
```

```python
plt.plot("Age","SibSp",data=df1)
plt.show()
```

```python
sns.boxplot(x="Pclass",y="Fare",data=df1)
plt.title("Fare by Pclass")
plt.show()
```
